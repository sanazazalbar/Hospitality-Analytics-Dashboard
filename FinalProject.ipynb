```python
df.drop_duplicates(inplace=True)
```

```python
df["Guest name"] = df["Guest name"].str.strip()
```

```python
df["Guest name"] = df["Guest name"].str.title()
df
```

```python
df["Room type"] = df["Room type"].replace({
    "STD": "Standard",
    "standard": "Standard",
    "suite ": "Suite",
    "SUTE": "Suite",
    "delux": "Deluxe",
    "DLX": "Deluxe"
})
```

```python
df["Room type"].unique()
```

```python
df["Service used"].unique()
```

```python
df["Service used"] = df["Service used"].replace({
    "spa": "SPA",
    "room service": "Room Service"
})
```

```python
df["Location"].unique()
```

```python
df["Booking channel"].unique()
```

```python
df["Booking channel"] = df["Booking channel"].replace({
    "website": "Website"
})
```

```python
df["Booking channel"].unique()
```

```python
df["Status"].unique()
```

```python
df["Status"] = df["Status"].replace({
    "confirmed": "Confirmed"
})
```

```python
df["Guest rating"] = df["Guest rating"].fillna(df["Guest rating"].mean())
```

```python
df["Review category"] = pd.cut(
    df["Guest rating"],
    bins=[0, 2.5, 3.5, 4.5, 5],
    labels=["Poor", "Average", "Good", "Excellent"]
)
```

```python
df["Review category"].isnull().sum()
```

```python
df["Room rating"].isnull().sum()
```

```python
df["Room rating"] = df["Room rating"].fillna(df["Room rating"].mean())
```

```python
df["Total revenue"].isnull().sum()
```

```python
df["Total revenue"] = df["Total revenue"].fillna(
    df["Room rating"] * df["Night stayes"]
)
```

```python
df["Total revenue"].isnull().sum()
```

```python
df["Month year"].isnull().sum()
```

```python
df["Operating cost"].isnull().sum()
df["Profit"].isnull().sum()
```

```python
df["Operating cost"] = df["Operating cost"].fillna(
    df["Operating cost"].median()
)
```

```python
df["Profit"] = (
    df["Total revenue"] - df["Operating cost"]
)
```
