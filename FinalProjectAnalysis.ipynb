```python
df["Month"] = pd.to_datetime(df["Check in date"]).dt.month_name()
```

```python
df["Year"] = pd.to_datetime(df["Check in date"]).dt.year
```

```python
df["Weekday"] = pd.to_datetime(df["Check in date"]).dt.day_name()
```

```python
total_bookings = df["Booking id"].count()

print("Total Bookings:", total_bookings)
```

```python
total_revenue = df["Total revenue"].sum()

print("Total Revenue:", total_revenue)
```

```python
avg_rating = df["Guest rating"].mean()

print("Average Guest Rating:", round(avg_rating,2))
```

```python
avg_room_rate = df["Room rating"].mean()

print("Average Room Rate:",round(avg_room_rate,2))
```

```python
ax = df.groupby("Location")["Total revenue"].sum().plot(kind='bar')

ax.bar_label(ax.containers[0])

plt.title("Revenue by Location")
plt.xlabel("Location")
plt.ylabel("Revenue")

plt.show()
```

```python
df["Status"].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title("Booking Status Distribution")

plt.show()
```

```python
df.groupby("Month")["Total revenue"].sum().plot(
    kind='line',
    marker='o'
)

plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")

plt.show()
```

```python
df["Room type"].value_counts().plot(kind='bar')
ax.bar_label(ax.containers[0])
plt.title("Bookings by Room Type")

plt.show()
```

```python
df["Guest rating"].plot(kind='hist')

plt.title("Guest Rating Distribution")

plt.show()
```

```python
df.boxplot(column="Total revenue")

plt.title("Revenue Outliers")

plt.show()
```

```python
Q1 = df["Total revenue"].quantile(0.25)

Q3 = df["Total revenue"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR

upper = Q3 + 1.5 * IQR

outliers = df[
    (df["Total revenue"] < lower) |
    (df["Total revenue"] > upper)
]

print(outliers)
```
